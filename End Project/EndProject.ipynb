## Cleaning the Data
* columns:</n>

```python
# Cleaning the genre column
# created new columns for genre id and genre name

df['genres'].head(10)
```

```python
import ast
```

```python
##extracting the id values from the list
genre_id = [[x["id"] for x in ast.literal_eval(movie)] for movie in df['genres']]
genre_id
```

```python
#extract the genre names
genre_name = [[x['name'] for x in ast.literal_eval(movie)] for movie in df['genres']]
df['genre_name']= genre_name
df.head(10)
```

```python
df['genre_name'][0]
```

```python
#changing the list to a string
# found on StackOverflow: 
#https://stackoverflow.com/questions/45306988/column-of-lists-convert-list-to-string-as-a-new-column

df['genre_names'] = [','.join(map(str,l)) for l in df['genre_name']]
```

```python
df['genre_names'][0]
```

```python
df['keywords'].head()
```

```python
# extract the keywords from the keywords column
keywords = [[x['name'] for x in ast.literal_eval(words)] for words in df['keywords']]
keywords
```

```python
#adjust the data frame
df['keywords'] = keywords
```

```python
df[['keywords']].head() # see the lists have been adjusted for
```

```python
#changing the keywords to a string
df['keywords'] = [','.join(map(str,l)) for l in df['keywords']]
```

```python
#check has been changed
df[['keywords']].head()
```

```python
# correcting production_companies
df['production_companies']
```

```python
# Correcting production_companies
df['production_companies']
```

```python
companies = [[x['name'] for x in ast.literal_eval(companies)] for companies in df['production_companies']]
```

```python
companies
```

```python
#adjust in data frame
df['production_companies'] = companies
df[['production_companies']].head()
```

```python
#changing to a string
df['production_companies'] = [','.join(map(str,c)) for c in df['production_companies']]
```

```python
df[['production_companies']].head()
```

```python
#correcting countries
countries = [[x['name'] for x in ast.literal_eval(country)] for country in df['production_countries']]
```

```python
#changing in the data frame
df['production_countries'] = countries
```

```python
# changing to a string
df['production_countries'] = [','.join(map(str,p)) for p in df['production_countries']]
```

```python
df[['production_countries']].head()
```

```python
# correcting the spoken_languages column
df[['spoken_languages']].head()
```

```python
#extracting the languages
languages = [[x['name'] for x in ast.literal_eval(language)] for language in df['spoken_languages']]
```

#### Does the Budget for a movie affect its popularity?

Expected: A higher budget would mean more input into the movie resulting in a higher quality in production, increasing the value and enjoymet and the popularity of the movie  
Result: Higher budget does not mean higher revenue or popularity

```python
#getting highest rated movies and sorting according to budget and popularity
movies = df[['title','budget','runtime','genre_names','popularity','revenue']].sort_values(by= ['popularity','budget'],ascending = False).copy()
# movies.head(10)
```

```python
#Finding the top 10 most popular movies in the df
top_10 = movies.head(10).sort_values('popularity',ascending=False ).copy()
# top_10
```

```python
top_10.dtypes
```

Below the table is sorted according to popularity.  
The trend for the top 10 movies is; the graph shows is that the budget does not nesessarily affect neither the popularity or the revenue earned by the movie.
Minions is the most popular movie and had the lowest budget.

```python
# Graph top 10 most popular movies budgets vs revenue
top_10 = top_10.sort_values(by='popularity') # sort in order of popularity so graph can show insights

plt.figure(figsize = (10,7))

sns.barplot(x = 'title' , ## column names
            y = 'budget', ## bar height
            data = top_10,
           palette = 'mako') ## data_frame! 

#X Labels
plt.xticks(rotation = 45,
            horizontalalignment = 'right',
          fontsize = 'small',
          fontweight ='medium')
plt.xlabel('Movie Title')

#y Labels
plt.ylabel('Budget for Movies',
            fontweight = 'medium')
plt.yticks(fontsize ='small')


#Title
plt.title('Budget for top 10 movies')

#Legend
# plt.legend(top_10['title'].sort_values('popularity', ascending=False),
#             fontsize = 'small',
#           loc= 'upper right',
#           fancybox = False)

plt.show()
```

#### Does the number of languages per movie affect popularity?

#### What affect does Release date have on the popularity of a movie?

```python
import datetime
```

```python
#changing the data type
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
#does release date affect popularity?
top_50['release_date'] = df['release_date'].copy()
# top_10.sort_values('popularity', ascending = False)
```

```python
# extract the year and save in new column
top_50['year'] = df['release_date'].dt.year
```

```python
#Popularity of movies by Release Date
plt.figure(figsize = (20,9))
date_group = top_50.groupby('year')[['popularity']].\
            mean().\
            rename(columns={'popularity': 'Average Popularity'}).\
            reset_index(drop=False)


sns.displot(x = date_group['year'] ,
              data = date_group,
            height=8,
           bins=100)

plt.xlabel('Year')
plt.ylabel('Popularity')

plt.xticks(rotation = 45,
           horizontalalignment = 'right',
           fontweight = 'light',
           fontsize = 'large',
          )
plt.yticks(label = date_group['Average Popularity']
            )

plt.title('Average Popularity by Release date')

plt.show()
```

```python
date_group.sort_values('Average Popularity', ascending= False)
```
